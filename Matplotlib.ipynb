```python
##Matplotlib Bars
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["x", "y", "z"])
y = np.array([6, 2, 10])
plt.bar(x,y)
plt.show()
```

```python
## Matplotlib Pie chart
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
x = ["Bmw", "Audi", "Jeep", "volvo"]
plt.pie(y, labels = x)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
x = ["Bmw", "Audi", "Jeep", "volvo"]
xexplode = [0.2, 0.2, 0.2, 0.2]
plt.pie(y, labels = x,explode = xexplode)
plt.show() 
```
