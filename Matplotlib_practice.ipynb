```python
import matplotlib.pyplot as plt

import numpy as np

y = np.array([35, 25, 25, 15])

mylabels = ["Python", "Maths", "English", "chemistry"]

myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)

plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([50, 40, 30, 20])
mylabels = ["CSE", "ECE", "Civil", "Mech"]
plt.pie(y, labels = mylabels)
plt.legend(title = "For subjects:")
plt.show() 
```
